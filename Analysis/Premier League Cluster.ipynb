```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

%matplotlib inline

pd.options.display.max_columns = None
```

```python
import pandas as pd
pl = pd.read_csv("C:\\Users\\\mushr\\Desktop\\433\\England18-19.csv",encoding = 'GBK',engine='python')
```

```python
# 胜负平处理

def judge_points(df):
    if df['goals'] > df['goalsLost']:
        df['points'] = 3
    elif df['goals'] == df['goalsLost']:
        df['points'] = 1
    else:
        df['points'] = 0
    return df['points']

pl['points'] = pl.apply(judge_points,axis=1)
```

```python
pl
```

```python
plt.scatter(pl.index,pl.passes)
```

```python
#转换成numpy array
X = pl[['passes','points']]

model = KMeans(n_clusters=5) #创建模型
model = model.fit(X) #训练模型
target = model.predict(X) #对原始数据进行标注
```

```python
pd.crosstab(target,target) #交叉表查看各个类别数据的数量
```

```python
plt.scatter(X.passes,X.points,c=target)#查看聚类的分布情况
```
